Evaluate missing data

```python
datasets = [kiva_loans, mpi_df, loan_theme_ids, loan_themes_by_region]
dataset_names = ['kiva_loans', 'mpi_df', 'loan_theme_ids', 'loan_themes_by_region']
```

```python
def find_null_values(df, name):
    print(f"\n--- Null values in {name} ---")
    total_rows = len(df)
    null_counts = df.isnull().sum()
    null_percent = (null_counts / total_rows) * 100
    null_summary = pd.DataFrame({
        'Missing Values': null_counts,
        'Percent Missing': null_percent.round(2)
    })
    print(null_summary[null_summary['Missing Values'] > 0]
          .sort_values(by='Missing Values', ascending=False))
```

```python
for df, name in zip(datasets, dataset_names):
    find_null_values(df, name)
```

kiva_loans is missing 8.46% of region values which is significant considering we're estimnating regional poverty levels. To handle this missing data, I will infer missing regions using available context:

step 1: check for a strong pattern between partner_ids and region  
step 2: use that mapping to fill missing regions  
step 3: drop the remaining rows with missing region

```python
region_map = kiva_loans.dropna(subset=['region']).groupby(['country', 'partner_id'])['region'].agg(lambda x: x.mode()[0]).reset_index()

def fill_region(row):
    if pd.isna(row['region']):
        match = region_map[
            (region_map['country'] == row['country']) & 
            (region_map['partner_id'] == row['partner_id'])
        ]
        if not match.empty:
            return match.iloc[0]['region']
    return row['region']

kiva_loans['region'] = kiva_loans.apply(fill_region, axis=1)
kiva_loans = kiva_loans.dropna(subset=['region'])
print(kiva_loans.isnull().sum().sort_values(ascending=False))
```

```python
import pandas as pd
import numpy as np
import re
from sklearn.feature_extraction.text import CountVectorizer

kiva_loans = kiva_loans.copy()

kiva_loans['use'] = kiva_loans['use'].fillna('')

def clean_text(text):
    text = text.lower()                         # Lowercase
    text = re.sub(r'[^a-z\s]', '', text)        # Remove punctuation and numbers
    return text

kiva_loans['clean_use'] = kiva_loans['use'].apply(clean_text)

vectorizer = CountVectorizer(
    stop_words='english',       
    max_features=1000           
)

X = vectorizer.fit_transform(kiva_loans['clean_use'])

bow_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

bow_df.head()
```

Here, I preprocessed the "use" column of the "kiva_loans" dataset to prepare it for analysis. I started by filling in missing values with empty strings and cleaning the text by converting it to lowercase and removing punctuation and numbers. Next, I used CountVectorizer to transform the cleaned text into a numerical format using the Bag-of-Words model. This resulted in a new DataFrame bow_df  containing word count features for the top 1,000 most frequent words, which can be used for further analysis or modeling.
